# Understanding of Image Data for Deep Learning

image.png

Consider a black & white or grey scale image. This is an image of number 8. Now if we keep zooming it we will see image is getting distorted and we will see small squares on the image. Now these are called as pixels. Also we use the term dimension of the image is X x Y which means nos. of pixels across height and width. Although we see the image in this format, the computer stores it in form of numbers where each of pixels are denoted as numerical value and this numerical value denotes the intensity of the pixel.

image.png

For grey-scale image intensity ranges from 0–255. Value closer to 0 represent darker shade where as value closer to 255 represent lighter shade. **So, every image in a computer is stored in a form of matrix of numbers and this matrix is also known as a channel.**
